## Matplotlib Plotting

```python
import matplotlib.pyplot as plt
import numpy as np
xpoints = np.array([-23, 38])
ypoints = np.array([-53, 40])
plt.plot(xpoints, ypoints)
plt.show()
```

## Matplotlib Bars

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["a", "b", "c", "d"])
y = np.array([5, 10, 3, 15])
plt.bar(x,y)
plt.show()
```

## Matplotlib Histograms

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.normal(150, 40, 350)
plt.hist(x)
plt.show() 
```

## Matplotlib Pie Charts

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([45, 25, 25, 5])
mylabels = ["cars", "tractors", "bikes", "autos"]
plt.pie(y, labels = mylabels)
plt.legend()
plt.show() 
```

## Pie Charts

## colors

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35, 25, 25, 15])
mylabels = ["mangoes", "carrots", "kevi", "grapes"]
mycolors = ["yellow", "orange", "brown", "green"]
plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

## shadow

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.4, 0, 0, 0]
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.3, 1, 0.1, 0.2]
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

## Horizontal Bars
